```python
# pytz.country_timezones.values()
local_tz = pytz.timezone('Australia/Hobart')
print local_tz.tzname
```

```python
# class datetime.datetime(year, month, day[, hour[, minute[, second[, microsecond[, tzinfo]]]]])¶
# df['Datetime'] = df.apply(lambda row: datetime.datetime(row['Year'], row['Month'], row['Day']), axis=1)
# df['LocalDateTime_mmddyyyy'] = df.apply(lambda row: datetime.datetime(row['Year Month Day Hour Minutes in YYYY'], row['MM'], row['DD'], row['HH24'], tzinfo=local_tz), axis=1)
df['LocalDateTime_mmddyyyy_unaware'] = df.apply(lambda row: datetime.datetime(row['Year Month Day Hour Minutes in YYYY'], row['MM'], row['DD'], row['HH24'],0,0, tzinfo=None), axis=1)
df['LocalDateTime_mmddyyyy_AEST'] = df.apply(lambda row: datetime.datetime(row['Year Month Day Hour Minutes in YYYY'], row['MM'], row['DD'], row['HH24'],0,0, tzinfo=local_tz), axis=1)
```

```python
# Localize the index (using tz_localize) to UTC (to make the Timestamps timezone-aware) 
# and then convert to Eastern (using tz_convert):
# df.LocalDateTime_mmddyyyy.tz_localize('Australia/Hobart')    
df.set_index(df.LocalDateTime_mmddyyyy_unaware, inplace=True)
```
