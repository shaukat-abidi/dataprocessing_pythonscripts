```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_src = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_1\\sampled_each_hour\\WeatherData_loc1_nov2016.csv")
df_dst = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_1\\sampled_each_hour\\WeatherData_loc1_nov2016_filtered.csv")
```

```python
# first convert UtcTime object to datetime
df_src.UtcTime = pd.to_datetime(df_src.UtcTime)
df_dst.UtcTime = pd.to_datetime(df_dst.UtcTime)
```

```python
print(df_src.UtcTime.dtype)
print(df_dst.UtcTime.dtype)
```

```python
# Generate date with hourly gaps
start_date = '2016-11-01'
end_date = '2016-11-28'
date_range = pd.date_range(start_date, end_date, freq='H')
```

```python
# print(date_range[0:-1])
# delete the last entry 
date_range = date_range[0:-1]
print(date_range)
```

```python
# set index for df
df_src = df_src.set_index(df_src.UtcTime)
df_dst = df_dst.set_index(df_dst.UtcTime)
```

```python
print('Total entries in date_range: %d \n' %(len(date_range)))
new_df = pd.DataFrame(columns=df_dst.columns) # Generate new dataframe
tot_entries = len(date_range)

# for loop
for entry in range(0,tot_entries):
    if (df_dst.index[entry] is pd.NaT):
        print('index: %d is missing, search in df_src: ----->> %s' %(entry, date_range[entry]))
        missing_date = date_range[entry]
        bool_value = missing_date in df_src.index
        min_offset = 1
        while bool_value is False:
            new_date = missing_date + pd.DateOffset(minutes=min_offset)
            min_offset = min_offset + 1
            bool_value = new_date in df_src.index
            print('searching %s in df_src\n' %(new_date) )
            if (bool_value is True):
                new_df.loc[entry] = df_src.loc[new_date]
                print('%s from df_src is inserted into new_df\n' %(new_date) )
    else:
        print('index: %d is available, therfore copying everything: ----- %s\n' %(entry, date_range[entry]))
        new_df.loc[entry] = df_dst.loc[date_range[entry]]
        
```
