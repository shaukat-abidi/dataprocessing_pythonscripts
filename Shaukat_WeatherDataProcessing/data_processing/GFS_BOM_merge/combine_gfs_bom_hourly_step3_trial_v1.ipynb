```python
import pandas as pd
import numpy as np

df_aug = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_5\\sampled_each_hour\\WeatherData_loc5_aug2016.csv")
df_sep = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_5\\sampled_each_hour\\WeatherData_loc5_sep2016.csv")
df_oct = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_5\\sampled_each_hour\\WeatherData_loc5_oct2016.csv")
df_nov = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\data_respository\\loc_5\\sampled_each_hour\\WeatherData_loc5_nov2016.csv")
df_gfs = pd.read_csv("C:\\Shaukat\\code\\data_rep\\gfs\\dunalley\\csv\\gfs_data_hourly.csv")
```
