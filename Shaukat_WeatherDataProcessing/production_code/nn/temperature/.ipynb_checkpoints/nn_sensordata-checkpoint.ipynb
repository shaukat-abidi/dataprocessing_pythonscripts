```python
#################
# Import Modules
#################
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from keras.callbacks import ModelCheckpoint, CSVLogger, ReduceLROnPlateau
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from keras.models import load_model

###########################
# For plots inside notebook
###########################
%matplotlib inline

#######################
# Impute sum of squares
#######################
def sum_of_square(vec_a, vec_b):
    diff = vec_a - vec_b
    diff_squared = np.square(diff)
    return (np.sum(diff_squared))
```

```python
#################################################################
# returns groundtruth and prediction vector for the next 168 hours
##################################################################
def generate_pred_gt_vec(test_ex,x_test_minmax,model,y_test_nn,df_test_for_eval,scaler_minmax):
    # At t
    # x_test_ex: Temperature_at_t Temperature_t-1 Temperature_t-2 pred_temp_t y_gfs_fcst_1

    # At t=t+1
    # x_test_ex: Temperature_t+1 Temperature_t Temperature_t-1 y_gfs_fcst_t y_gfs_fcst_t+1

    pred_list = []
    x_test_ex_arr = np.arange(5.0).reshape(1,5)


    pred_val = model.predict(x_test_minmax)[test_ex]
    pred_list.append(pred_val)

    x_test_dim1_lasttime = df_test_for_eval.ix[test_ex,"Temperature"]
    x_test_dim2_lasttime = df_test_for_eval.ix[test_ex,"Temperature_past1hr"]
    x_test_dim3_lasttime = df_test_for_eval.ix[test_ex,"Temperature_past2hr"]
    x_test_dim4_lasttime = df_test_for_eval.ix[test_ex,"pred_temp"]
    x_test_dim5_lasttime = df_test_for_eval.ix[test_ex,"y_gfs_fcst_1"]

    next_hours = 168 # Next seven days (168 hours)
    for hr in range(1,next_hours):
        hr_t_1 = hr + 1
        col_name_gfs_1 = 'y_gfs_fcst_'+str(hr)
        col_name_gfs_2 = 'y_gfs_fcst_'+str(hr_t_1)

        # prepare test instance the next-hour forecast
        x_test_dim1_current = pred_val
        x_test_dim2_current = x_test_dim1_lasttime
        x_test_dim3_current = x_test_dim2_lasttime
        x_test_dim4_current = df_test_for_eval.ix[test_ex,col_name_gfs_1] 
        x_test_dim5_current = df_test_for_eval.ix[test_ex,col_name_gfs_2] 

        # Populate numpy array
        x_test_ex_arr[0,0] = x_test_dim1_current
        x_test_ex_arr[0,1] = x_test_dim2_current
        x_test_ex_arr[0,2] = x_test_dim3_current
        x_test_ex_arr[0,3] = x_test_dim4_current
        x_test_ex_arr[0,4] = x_test_dim5_current

        # Transform it
        x_test_ex_arr_transformed = scaler_minmax.transform(x_test_ex_arr)

        # Predict and add it to the list
        pred_val = model.predict(x_test_ex_arr_transformed)
        pred_list.append(pred_val)

        # Store current vector
        x_test_dim1_lasttime = x_test_dim1_current
        x_test_dim2_lasttime = x_test_dim2_current
        x_test_dim3_lasttime = x_test_dim3_current
        x_test_dim4_lasttime = x_test_dim4_current
        x_test_dim5_lasttime = x_test_dim5_current

    y_pred_vec = np.asarray(pred_list)
    y_gt_vec = y_test[test_ex,:]
    return y_gt_vec,y_pred_vec

```

```python
######################
# Read sensor df here
######################
df_sensor = pd.read_csv("C:/Shaukat/code/data_rep/sensor_data/Pooleys_05062017_temperature.csv")

#####################
# Read GFS data here
######################
df_gfs = pd.read_csv("C:/Shaukat/code/data_rep/gfs/2016_2017/csv/pooley_1hr.csv")

###########################
# Set the training ratio 
###########################
training_ratio = 0.7
```

```python
df_gfs = df_gfs
```

```python
#########################
# Setting up sensor data
########################

# Sensor Time is in UTC
df_sensor.EndTime = pd.to_datetime(df_sensor.EndTime)
df_sensor.set_index('EndTime', inplace=True)


# Initialize empty df
df_sensor_proc = pd.DataFrame(index=df_sensor.index)
df_sensor_proc['Temperature'] = df_sensor['Average']

# Get start and end dates of data stream
start_date = df_sensor_proc.index[0]
end_date = df_sensor_proc.index[len(df_sensor_proc)-1]

# start_date and end_date should end in xx:xx:00 format so that they can be aligned with GFS
start_date = start_date - pd.Timedelta(minutes = start_date.minute, seconds = start_date.second)
end_date = end_date - pd.Timedelta(minutes = end_date.minute, seconds = end_date.second)

# print dates
print 'start_date: ',start_date
print 'end_date: ',end_date

#initialize dateranges for hourly dataframe
_daterange = pd.date_range(start_date,end_date,freq='H')
print _daterange

# Generate hourly df
tot_entries = len(_daterange)
col_list = df_sensor_proc.columns.tolist()
print 'Columns for df_sensor_proc: ',col_list
df_sensor_proc_hourly = pd.DataFrame(index = _daterange, columns=col_list)

#######################
# df_sensor_proc_hourly
#######################
for ent in range(1,tot_entries):
    current_index = _daterange[ent]

    # copy the corresponding patch of dataframe
    df_patch = df_sensor_proc.loc[_daterange[ent-1] + pd.Timedelta(minutes=1):_daterange[ent]].copy()

    # Fill the corresponding entry of dataframe
    # DataFrame.set_value(index, col, value, takeable=False)
    df_sensor_proc_hourly.set_value(current_index,'Temperature',df_patch.Temperature.mean())

# Drop the first row
df_sensor_proc_hourly.drop(df_sensor_proc_hourly.index[0], inplace = True)
```

```python
#####################
# Setting up GFS
#####################

# Select the desired columns of GFS
df_gfs = df_gfs[ ['pred_temp', 'Unnamed: 0'] ]

# Change the unit of temperature
df_gfs.pred_temp = df_gfs.pred_temp - 273.15 # Kelvin to Celsius

# Setting up GFS UTC time as index
df_gfs['Unnamed: 0'] = pd.to_datetime(df_gfs['Unnamed: 0'])
df_gfs.set_index('Unnamed: 0', inplace = True)
```

```python
###########################################
# combine df_gfs and df_sensor_proc_hourly
###########################################
df_gfs_bom_combined = df_sensor_proc_hourly.merge(df_gfs, how='inner', left_index=True, right_index=True)
print len(df_gfs_bom_combined)
df_gfs_bom_combined.dropna(axis=0, how='all', inplace = True) # Drop the rows where all of the elements are nan
print len(df_gfs_bom_combined)

# interpolate df_gfs_bom_combined
for col in df_gfs_bom_combined:
    print col
    df_gfs_bom_combined[col] = pd.to_numeric(df_gfs_bom_combined[col])
    
df_gfs_bom_combined = df_gfs_bom_combined.interpolate(method='time')

#######################
# Important step
########################
df = df_gfs_bom_combined.copy()
```

```python
###################################
# Prepare train/test dataframe here
###################################

# Generate df for gt label (Y)
df_gt_output = pd.DataFrame(index = df.index)
df_gt_output['Temperature'] = df.Temperature.copy()

# Make a copy of df for data matrix (X)
df_data = df.copy()

# Add groundtruth temp for the next 7-days from sensordata
for hrs in range(1,169):
    col_name = 'y_t_'+str(hrs)
    df_gt_output[col_name] = df_gt_output.Temperature.shift(-hrs)
    # print col_name

# Now remove the current temperature from this dataframe
df_gt_output = df_gt_output.drop('Temperature', axis=1)

# Get the observation from past two hours from sensordata
past_hours = 3

# BOM Temperature for the last two hours
current_col = 'Temperature'
for hr in range(1,past_hours):
    col_name = current_col+'_past'+str(hr)+'hr'
    df_data[col_name] = df[current_col].shift(hr) # Include past values
    #print col_name
    
# Add GFS forecasts into the input for the next hour
current_col = 'pred_temp'
next_hours = 169
for hr in range(1,next_hours):
    col_name = 'y_gfs_fcst_'+str(hr)
    df_data[col_name] = df[current_col].shift(-hr) # Include Future Values from GFS
    #print col_name

tot_cols_df_data = len(df_data.columns)
# Combine data and temp forecast for the next seven days and filter nan
df_combined = pd.concat([df_data,df_gt_output], axis=1)
df_combined = df_combined.dropna()

# Get the list of column names
data_columns = df_combined.columns[0:tot_cols_df_data]
gt_temp_columns = df_combined.columns[tot_cols_df_data:]

# create new dataframe for data and forecast
df_data_nan_filtered = df_combined[data_columns]
df_gt_output_nan_filtered = df_combined[gt_temp_columns]

# Create df for NN
df_data_nn = pd.DataFrame()
df_data_nn['Temperature'] = df_data_nan_filtered.Temperature.copy()
df_data_nn['Temperature_past1hr'] = df_data_nan_filtered.Temperature_past1hr.copy()
df_data_nn['Temperature_past2hr'] = df_data_nan_filtered.Temperature_past2hr.copy()
df_data_nn['pred_temp'] = df_data_nan_filtered.pred_temp.copy()
df_data_nn['y_gfs_fcst_1'] = df_data_nan_filtered.y_gfs_fcst_1.copy()
# df_data_nn.head(5)
df_data_matrix = df_data_nn.as_matrix()
df_gt_temp_forecast_matrix = df_gt_output_nan_filtered.as_matrix()

# Prepare train and test split
tot_points = df_data_matrix.shape[0]
train_points = int(np.floor(tot_points*training_ratio))
test_points = tot_points - train_points
print 'tot_points: %d, train_points: %d, test_points:%d' %(tot_points, train_points, test_points)

# Generate Train Sequence
x_train = df_data_matrix[0:train_points + 1,:]
y_train = df_gt_temp_forecast_matrix[0:train_points + 1,:]
# Generate Test Sequence
x_test = df_data_matrix[train_points + 1:tot_points + 1 ,:]
y_test = df_gt_temp_forecast_matrix[train_points + 1:tot_points + 1 ,:]

print 'x_train_shape: ',x_train.shape
print 'y_train_shape: ',y_train.shape
print 'x_test_shape: ',x_test.shape
print 'y_test_shape: ',y_test.shape

# labelings for nn
y_train_nn = np.array(y_train[0:,0:1]).copy()
y_test_nn = np.array(y_test[0:,0:1]).copy()
print 'y_train_for_nn_shape: ',y_train_nn.shape
print 'y_test_for_nn_shape: ',y_test_nn.shape

# Do scaling according to min-max scaling
scaler_minmax = preprocessing.MinMaxScaler()
x_train = scaler_minmax.fit_transform(x_train)
x_test_minmax = scaler_minmax.transform(x_test)

print 'x_test_minmax_shape: ',x_test_minmax.shape
```

```python
print 'train_points: ',train_points,'test_points: ',test_points,'tot_points: ',tot_points

# Generate Train DF
df_train_for_eval = df_data_nan_filtered.iloc[0:train_points + 1,:]

# Generate Test DF
df_test_for_eval = df_data_nan_filtered.iloc[train_points + 1:tot_points + 1 ,:]
```

```python
##################
# Learn the model
##################
model = Sequential()
model.add(Dense(500, input_dim=x_train.shape[1], init='normal', activation='relu')) #1st Hidden Layer
model.add(Dense(250, init='normal', activation='relu')) #next Hidden Layer
model.add(Dense(100, init='normal', activation='relu')) #next Hidden Layer
model.add(Dense(y_train_nn.shape[1], init='normal')) #output layer

################
# compile model
################
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
###########################################
# Setting up callbacks and fire-up learning
###########################################
filepath_savemodel = 'C:\\Shaukat\\code\\temperature_models\\models\\sensor_data\\s8\\model_s8_best.h5'
filepath_csvlogger = 'C:\\Shaukat\\code\\temperature_models\\models\\sensor_data\\s8\\training_s8.log'

checkpointer = ModelCheckpoint(filepath_savemodel, monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=False, mode='auto')
csv_logger = CSVLogger(filepath_csvlogger, separator=',', append=False)

history = model.fit(x_train, y_train_nn, nb_epoch=100, verbose=1, validation_data=(x_test_minmax, y_test_nn), batch_size = 32, shuffle=True, callbacks=[checkpointer, csv_logger])

```
