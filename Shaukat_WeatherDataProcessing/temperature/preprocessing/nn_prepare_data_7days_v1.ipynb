```python
df_gfs = pd.read_csv("C:\\Users\\ShaukatAbidi\\Documents\\shaukat_python_progs\\eto_experiment\\data_eto\\gfs_data\\gfs\\gfs_combined.csv")
```

```python
df_gfs.set_index('Unnamed: 0', inplace=True)
```

```python
df_gfs.index = pd.to_datetime(df_gfs.index)
```

```python
df_gfs.index
```

```python
# Generate date with hourly gaps
start_date = '2016-06-01'
end_date = '2016-12-09'
date_range = pd.date_range(start_date, end_date, freq='H')
```

```python
df_gfs_hourly = pd.DataFrame(index=date_range, columns=df_gfs.columns)
```

```python
df_gfs_hourly
```

```python
# Copy df_gfs into df_gfs_hourly
tot_len = len(df_gfs)
# ind =  df_gfs.index[1]
# print ind, len(df_gfs)

for entry in range(0,tot_len):
    ind = df_gfs.index[entry]
    df_gfs_hourly.loc[ind] = df_gfs.loc[ind]

# df_gfs_hourly.loc[ind]
```

```python
df_gfs_hourly
```

```python
df_gfs_hourly
```

```python
# For interpolation, fill the first row of df_gfs_hourly
df_gfs_hourly.iloc[0] = df_gfs_hourly.iloc[3]
```

```python
df_gfs_hourly.describe()
```

```python
# df_gfs_hourly['pred_cloud_cover']
for col in df_gfs_hourly:
    # print col
    df_gfs_hourly[col] = pd.to_numeric(df_gfs_hourly[col])
```

```python
# df_gfs_hourly.dtypes
# now interpolate
df_gfs_hourly = df_gfs_hourly.interpolate(method='time')
```
